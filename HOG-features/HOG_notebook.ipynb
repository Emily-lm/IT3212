```python
# Test against scikit-image HOG

from skimage.io import imread, imshow
from skimage.transform import resize
from skimage.feature import hog
from skimage import exposure
import matplotlib.pyplot as plt


```

```python
resized_img = resize(optimus, (128, 128))
#resized_img = optimus

#creating hog features 
fd, hog_image = hog(resized_img, orientations=9, pixels_per_cell=(8, 8), 
                    cells_per_block=(2, 2), visualize=True, channel_axis=-1)

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(16, 8), sharex=True, sharey=True) 

ax1.imshow(resized_img, cmap=plt.cm.gray) 
ax1.set_title('Input image') 

# Rescale histogram for better display 
hog_image_rescaled = exposure.rescale_intensity(hog_image, in_range=(0, 10)) 
_, mag_normalized, _= hogg.hog(resized_img, (8,8), (2,2), 9)

ax2.imshow(mag_normalized, cmap=plt.cm.gray)
ax2.set_title('Gradient Image')

ax3.imshow(hog_image_rescaled, cmap=plt.cm.gray) 
ax3.set_title('Histogram of Oriented Gradients')

plt.show()
```

```python
resized_img = resize(starship, (128, 128))
#resized_img = bird

for i in range(3):
    fd, hog_image = hog(resized_img, orientations=9, pixels_per_cell=(2*(i+2),2*(i+2)), 
                    cells_per_block=(2, 2), visualize=True, channel_axis=-1)
    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(16, 8), sharex=True, sharey=True) 

    ax1.imshow(resized_img, cmap=plt.cm.gray) 
    ax1.set_title('Input image') 

    # Rescale histogram for better display 
    hog_image_rescaled = exposure.rescale_intensity(hog_image, in_range=(0, 10)) 
    _, mag_normalized, _= hogg.hog(resized_img, (2*(i+2),2*(i+2)), (2,2), 9)

    ax2.imshow(mag_normalized, cmap=plt.cm.gray)
    ax2.set_title('Gradient Image')

    ax3.imshow(hog_image_rescaled, cmap=plt.cm.gray) 
    ax3.set_title('Histogram of Oriented Gradients, cell size = ' + str(2*(i+2))), 'block size = ' + str((i+1))

    plt.show()
```

```python
resized_img = resize(truck, (128, 128))
fd1, hog_image1 = hog(resized_img, orientations=2, pixels_per_cell=(8, 8), 
                    cells_per_block=(2, 2), visualize=True, channel_axis=-1)

fd2, hog_image2 = hog(resized_img, orientations=4, pixels_per_cell=(8, 8), 
                    cells_per_block=(2, 2), visualize=True, channel_axis=-1)

fd3, hog_image3 = hog(resized_img, orientations=20, pixels_per_cell=(8, 8), 
                    cells_per_block=(2, 2), visualize=True, channel_axis=-1)


fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(16, 8), sharex=True, sharey=True) 


# Rescale histogram for better display 
hog_image_rescaled1 = exposure.rescale_intensity(hog_image1, in_range=(0, 10)) 
hog_image_rescaled2 = exposure.rescale_intensity(hog_image2, in_range=(0, 10))
hog_image_rescaled3 = exposure.rescale_intensity(hog_image3, in_range=(0, 10))

_, mag_normalized, _= hogg.hog(resized_img, (8,8), (2,2), 9)


ax1.imshow(hog_image_rescaled1, cmap=plt.cm.gray) 
ax1.set_title('HOG image' + " orientations: " + str(2)) 

ax2.imshow(hog_image_rescaled2, cmap=plt.cm.gray)
ax2.set_title('HOG image' + " orientations: " + str(3))


ax3.imshow(hog_image_rescaled3, cmap=plt.cm.gray) 
ax3.set_title('HOG image' + " orientations: " + str(20))
```
