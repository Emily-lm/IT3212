```python
# Imports
%load_ext autoreload
%autoreload 2
import LBP_methods
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
# load images
onsager = cv2.imread('images/Onsager.jpg', cv2.IMREAD_GRAYSCALE)
iceland = cv2.imread('images/iceland.jpg', cv2.IMREAD_GRAYSCALE)
texture = cv2.imread('images/texture.jpg', cv2.IMREAD_GRAYSCALE)

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))

ax1.imshow(onsager, cmap='gray')
ax1.set_title('Onsager')
ax1.axis('off')

ax2.imshow(iceland, cmap='gray')
ax2.set_title('Iceland')
ax2.axis('off')

ax3.imshow(texture, cmap='gray')
ax3.set_title('Texture')
ax3.axis('off')

plt.show()
```

```python
# LBP
onsager_LBP = LBP_methods.lbp(onsager)
iceland_LBP = LBP_methods.lbp(iceland)
texture_LBP = LBP_methods.lbp(texture)

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))

ax1.imshow(onsager_LBP, cmap='gray')
ax1.set_title('Onsager LBP')
ax1.axis('off')

ax2.imshow(iceland_LBP, cmap='gray')
ax2.set_title('Iceland LBP')
ax2.axis('off')

ax3.imshow(texture_LBP, cmap='gray')
ax3.set_title('Texture LBP')
ax3.axis('off')

plt.show()
```

```python
# Histograms

onsager_hist = LBP_methods.hist(onsager_LBP)
iceland_hist = LBP_methods.hist(iceland_LBP)
texture_hist = LBP_methods.hist(texture_LBP)

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))

ax1.bar(range(256), onsager_hist)
ax1.set_title('Onsager Histogram')
ax1.set_xlabel('Bins')
ax1.set_ylabel('Frequency')

ax2.bar(range(256), iceland_hist)
ax2.set_title('Iceland Histogram')
ax2.set_xlabel('Bins')
ax2.set_ylabel('Frequency')

ax3.bar(range(256), texture_hist)
ax3.set_title('Texture Histogram')
ax3.set_xlabel('Bins')
ax3.set_ylabel('Frequency')

plt.show()
```

```python
onsager_lpb = LBP_methods.compute_lbp(onsager, radius=1, n_points=8)

iceland_lbp = LBP_methods.compute_lbp(iceland, radius=1, n_points=8)


texture_lbp = LBP_methods.compute_lbp(texture, radius=1, n_points=8)

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))

ax1.imshow(onsager_lpb, cmap='gray')
ax1.set_title('Onsager LBP')
ax1.axis('off')

ax2.imshow(iceland_lbp, cmap='gray')
ax2.set_title('Iceland LBP')
ax2.axis('off')

ax3.imshow(texture_lbp, cmap='gray')
ax3.set_title('Texture LBP')
ax3.axis('off')

plt.show()
```

```python
# plot histograms
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))

ax1.hist(onsager_lpb.ravel(), bins=256, range=(0, 256))
ax1.set_title('Onsager Histogram')
ax1.set_xlabel('Bins')
ax1.set_ylabel('Frequency')

ax2.hist(iceland_lbp.ravel(), bins=256, range=(0, 256))
ax2.set_title('Iceland Histogram')
ax2.set_xlabel('Bins')
ax2.set_ylabel('Frequency')

ax3.hist(texture_lbp.ravel(), bins=256, range=(0, 256))
ax3.set_title('Texture Histogram')
ax3.set_xlabel('Bins')
ax3.set_ylabel('Frequency')

plt.show()

```

```python
# plot histograms with percentages
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))

# Onsager LBP histogram
counts, bins, _ = ax1.hist(onsager_lpb.ravel(), bins=256, range=(0, 256), density=True)
ax1.set_title('Onsager Histogram')
ax1.set_xlabel('Bins')
ax1.set_ylabel('Percentage')
ax1.bar(bins[:-1], counts * 100, width=bins[1] - bins[0], color='blue', align='edge')

# Iceland LBP histogram
counts, bins, _ = ax2.hist(iceland_lbp.ravel(), bins=256, range=(0, 256), density=True)
ax2.set_title('Iceland Histogram')
ax2.set_xlabel('Bins')
ax2.set_ylabel('Percentage')
ax2.bar(bins[:-1], counts * 100, width=bins[1] - bins[0], color='green', align='edge')

# Texture LBP histogram
# Remove last bin


counts, bins, _ = ax3.hist(texture_lbp.ravel(), bins=256, range=(0, 256), density=True)
ax3.set_title('Texture Histogram')
ax3.set_xlabel('Bins')
ax3.set_ylabel('Percentage')
ax3.bar(bins[:-1], counts * 100, width=bins[1] - bins[0], color='orange', align='edge')

plt.show()
```
